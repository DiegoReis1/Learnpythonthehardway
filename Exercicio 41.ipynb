```python
import random
from urllib import urlopen
import sys

URL_DE_PALAVRA = "http://learncodethehardway.org/words.txt"
PALAVRAS = []

FRASES = {
    "class %%%(%%%):":
      "Criar uma classe chamada %%%Aquilo é um %%%.",
    "class %%%(object):\n\tdef __init__(self, ***)" :
      "classe %%% tem um __init__ Que leva a si mesmo e *** parametros.",
    "class %%%(object):\n\tdef ***(self, @@@)":
      "classe %%% tem uma função chamada *** Que leva a si mesmo e @@@ parametros.",
    "*** = %%%()":
      "Conjunto *** para instanciar uma classe %%%.",
    "***.***(@@@)":
      "A partir de *** pegue a *** função, E chamá-lo com parâmetros auto, @@@.",
    "***.*** = '***'":
      "A partir de *** pegue o *** Atributo e configurá-lo para '***'."
}

# do they want to drill phrases first
if len(sys.argv) == 2 and sys.argv[1] == "english":
    FRASE_UM = True
else:
    FRASE_UM = False

# load up the words from the website
for palavra in urlopen(URL_DE_PALAVRA).readlines():
    PALAVRAS.append(palavra.strip())


def convert(snippet, frase):
    nome_classes = [w.capitalize() for w in
                   random.sample(PALAVRAS, snippet.count("%%%"))]
    outros_nomes = random.sample(PALAVRAS, snippet.count("***"))
    resultados = []
    nomes_paragrafos = []

    for i in range(0, snippet.count("@@@")):
        contador_paragrafos = random.randint(1,3)
        nomes_paragrafos.append(', '.join(random.sample(PALAVRAS, contador_paragrafos)))

    for sentenca in snippet, frase:
        resultado = sentenca[:]

        # fake class names
        for palavra in nome_classes:
            resultado = resultado.replace("%%%", palavra, 1)

        # fake other names
        for palavra in outros_nomes:
            resultado = resultado.replace("***", palavra, 1)

        # fake parameter lists
        for palavra in nomes_paragrafos:
            resultado = resultado.replace("@@@", palavra, 1)

        resultados.append(resultado)

    return resultados


# keep going until they hit CTRL-D
try:
    while True:
        snippets = FRASES.keys()
        random.shuffle(snippets)

        for snippet in snippets:
            frase = FRASES[snippet]
            questao, resposta = convert(snippet, frase)
            if FRASE_UM:
                questao, resposta = resposta, questao

            print questao

            raw_input("> ")
            print "RESPOSTA:  %s\n\n" % resposta
except EOFError:
    print "\nTchau"
```
